```python
def analyser(name):
    vowels = "aeiouAEIOU"
    vowel_count = sum(1 for char in name if char in vowels)
    consonant_count = sum(1 for char in name if char.isalpha() and char not in vowels)
    return vowel_count, consonant_count


print("Welcome to smart Username generator")
first_name = input("Enter your first name: ")
last_name = input("Enter your last name: ")
pet_name = input("Enter your pet name: ")

username = first_name[:3] + last_name[:3] + pet_name[:3]
print("Your generated username is: " , username)

l = first_name.upper()
print("First name in uppercase: " , l)
u = last_name.lower()
print("Last name in lowercase: " , u)
p = pet_name.capitalize()
print("Pet name with first letter capitalized: " , p)
smart_username = l + u + p
print("Your smart username is: " , smart_username)

print(f"reverse of your name is: {smart_username[::-1]}")

print(f"replace: {username.replace('a', '@')} spaces: {username.replace(' ', '_')}")

print(f"strip: {username.strip()}")
print(f"length: {len(username)}")

print("")
print("\n✅ Username generation complete!")

print("the Vowels and consonents are: ",analyser(smart_username))
```
