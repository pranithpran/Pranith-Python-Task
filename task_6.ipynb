```python
from datetime import datetime
task = []
while True:
    print("\n----- TO-DO LIST MANAGER Menu -----")
    print("1. Add Task")
    print("2. View Tasks")
    print("3. Exit")

    choice = int(input("Enter your choice (1-3): "))
    if choice ==1:
        add_task = input("please enter the task you want to add: ")
        task.append(add_task)
        print(f'Task added successfully: "{add_task}"')
        now = datetime.now()
        print("Current Time:", now)
    elif choice == 2:
        if not task:
            print("No tasks in the list.")
        else:
            print("your task")
            for i,t in enumerate(task,1):
                print(f"{i}.{t}")
    elif choice==3:
        print("Exiting the TO-DO LIST MANAGER. Goodbye!")
        break
    else:
        print("Invalid choice. Please enter a number between 1 and 3.")
```
