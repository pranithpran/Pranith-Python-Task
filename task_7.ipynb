```python
library ={}
running = True
while running:
    print("\n ===== PERSONAL LIBRARY MANAGEMENT =====")
    print("1. Add Book")
    print("2. Remove Book")
    print("3. Search Book")
    print("4. Update Book")
    print("5. Generate Report")
    print("6. Statistics")
    print("7. Exit\n")

    choice =int(input("enter your choice between 1-7: "))
    if choice ==1:
        print("Add Book")
        book_id = input("Enter Book id: ")
        if book_id in library:
            print("Book id already exist! ")
        else:
            title = input("Enter Title: ")
            author = input("Enter Author: ")
            year = int(input("Enter Published Year: "))
            genre = input("Enter Genre: ")
            pages = int(input("Enter Number of Pages: "))

            library[book_id] ={"title":title,
                               "author": author,
                               "year": year,
                               "genre": genre,
                               "pages":pages}
            print("Book added successfully!")
    elif choice ==2:
        print("Remove book")
        book_id = input ("enter book id to remove: ")
        if book_id not in library:
            print("book id not found! ")
        else:
            del library[book_id]
            print("Book id not found! ")


    elif choice ==3:
        print("search book by --> author/title/year/genere  ")
        field = input("enter field:author/title/year/genere ").lower()

        if field not in ['title','author','year','genre']:
            print("invalid field! ")

        else:
            value =input("enter value: ")
            if field == "year":
                value =int(value)   

            found = False

            for book_id in library:
                if library[book_id][field] == value:
                    print("\nfound: ", book_id, "-->", library[book_id])
                    found =True
            if not found:
                    print("no matching books found--> *** ")
    elif choice ==4:
        print("Update book: ")
        book_id = input("enter book_id to update: ")
        if book_id not in library:
            print(" Book not found! ")
        else:
            print("Book fields you can update are: title, author, year, genre, pages")
            field = input("enter field: ").lower()

            if field not in library[book_id]:
                print(" invalid field ! ")
            else:
                new_value = input("enter new value: ")
                if field in ["year", "pages"]:
                    new_value = int(new_value)

                library[book_id][field] = new_value
                print("Book updated successfully! ")
    elif choice ==5:
        print("\n ----------- LIBRARY REPORT -----------")

        for book_id in library:
            print(book_id, ":", library[book_id])
        print("--------------------------------------")

    elif choice == 6:
        print("Statistics: ")
        total_books =0
        total_pages = 0
        year_count = {}
        genre_count = {}

        for book_id in library:
            total_books +=1
            total_pages += library[book_id]["pages"]
            
            year = library[book_id]["year"]
            genre = library[book_id]["genre"]   

            if year not in year_count:
                year_count[year] =0
            year_count[year] +=1

            if genre not in genre_count:
                genre_count[genre] =0
            genre_count[genre] +=1

        print("\n ---------- STATISTICS ---------")
        print("Total Books:", total_books)
        print("Total Pages:", total_pages)
        print("Books per Year:", year_count)
        print("Books per Genre:", genre_count)
        print("--------------------------------")
    
    elif choice == 7:
        print("Exiting programme.. ")
        running = False
    else:
        print("Invalid choice! Try again. ")
        
```
