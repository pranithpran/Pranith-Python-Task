```python
students = []   # list to store all student records

#iterate 3 times to get student name and grade

for i in range(3):
    name = input("Enter student name: ")
    while True:
        grade_input = input("Enter grade: ")
        try:
            grade = int(grade_input)
            break
        except ValueError:
            print(f"Invalid input '{grade_input}'. Please enter an integer grade.")

    student = (name, grade)     # tuple (name, grade)
    students.append(student)    # add tuple to list

print("\nAll Student Grades:")
for student in students:
    print("Name:", student[0], "| Grade:", student[1])
```
